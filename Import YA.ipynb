```python
import yfinance as yf
import matplotlib.pyplot as plt

# What symbol take from Yahoo?
ticker_symbol = 'NVDA'  

# Loading 
brent_data = yf.download(ticker_symbol, start='2020-01-01', end='2023-01-01')

print(brent_data.head())

# Price plot
plt.figure(figsize=(10, 6))
plt.plot(brent_data['Close'], label='Close')
plt.title('Brent price')
plt.xlabel('Data')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
